## Kpop Merchandise Analysis

While there are plenty of kpop groups to choose from, it makes sense to compare "competing" groups for differences in the pricing of their products on a group level and on a store to store level. This is because while they may have distinct styles, presentation, and fanbases, these groups are competitors due to their size, fame, and relevance to their parent entertainment company. Hence I will compare the highest number of merchandise sold out and which platform do well for the artist.

### Questions

1. Artist with number of highest sold out merchandise in JYP
2. Artist with lowest number of merchandise sold out
3. Which platform have the highest sold out items

```python
#count duplicated data for each row
#create new variable to group the duplicate data 
# size () to indicate the index of duplicate element
data_dups = data.groupby(data.columns.tolist(), as_index=True).size()
display(data_dups)
```

```python
#remove duplicated data
data = data.drop_duplicates()
```

```python
#fill NaN values with 0
data.fillna(0)
```

### EDA with data

```python
data.describe(include = "all")
```

```python
#look at artist and sold out number
#use cross tab to create new table

soldout_tab = (pd.crosstab(data.loc[:, "artist"], 
                             data.loc[:, "is_sold_out"],
                             normalize = "index") * 100).round(2)
soldout_tab 
```

```python
soldout_tab = pd.crosstab(data.loc[:, "artist"], 
                             data.loc[:, "is_sold_out"])
                         
soldout_tab
```

```python
Figure1= sns.catplot(x= "is_sold_out", hue="artist", kind="count", data= data)
Figure1.fig.suptitle("Artist with Highest Merchandise Sold out", fontsize=12)
Figure1.set_axis_labels(x_var="Sold out merchandise", y_var="Number merchandise sold")
```

```python
Figure2= sns.catplot(x= "is_sold_out", hue="vendor", kind="count", data= data)
Figure2.fig.suptitle("Different Platform with Sold Merchandise", fontsize=12)
Figure2.set_axis_labels(x_var="Sold out merchandise", y_var="Number merchandise sold")
```
